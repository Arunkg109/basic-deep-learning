### Use the Dataset named HR Interview Analytics (given at the end) to predict the Status of the candidate (0 is not accepted and 1 is accepted).  Build an Artificial Neural Network to predict the Status. 

```python
# import libraries
import pandas as pd
import tensorflow as tf
```

```python
# import data

HR = pd.read_csv('F:\Skillenable\Data frames\HR-Interview-Analytics.csv')
HR.head()
```

```python
hr= HR.drop(columns=['SLNO','Candidate.Ref', 'Candidate.Source'])
hr.head()
```

```python
hr.info()
```

```python
# dealing with categorical col.

cat_col = ['DOJ.Extended', 'Offered.band', 'Joining.Bonus', 'Candidate.relocate.actual', 'Gender', 'LOB', 'Location']
hr_dummy = pd.get_dummies(hr, columns=cat_col, drop_first = True)
```

```python
hr_dummy.head()
```

```python
# split data into train and test
from sklearn.model_selection import train_test_split
Y = hr_dummy[['Status']]
X = hr_dummy.drop(columns=['Status'])

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)

len(x_train), len(x_test), len(y_train), len(y_test)
```

```python
# Building the ANN
## step 1. Initializing the ANN

ann = tf.keras.models.Sequential()   # ANN initialize 
```

```python
ann.add(tf.keras.layers.Dense(units= 10, activation='relu')) 
```

```python
ann.add(tf.keras.layers.Dense(units= 1, activation='sigmoid'))
```

```python
ann.compile(optimizer = 'adam', loss='binary_crossentropy', metrics = 'accuracy')
```

```python
ann.fit(x_train, y_train, batch_size=64, epochs= 100)  # training the data
```

```python
y_test
```

```python
y_test['Prediction'] = ann.predict(x_test)
```
