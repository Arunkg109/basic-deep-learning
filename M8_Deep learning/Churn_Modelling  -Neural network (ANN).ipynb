```python
# import data

churn = pd.read_csv('F:\Skillenable\Data frames\Churn_Modelling.csv')
churn.head()
```

```python
# dealing with categorical col.

cat_col = ['Geography', 'Gender']
churn_dummy = pd.get_dummies(churn, columns=cat_col, drop_first = True)
```

```python
churn_dummy.head()
```

```python
# split data into train and test
from sklearn.model_selection import train_test_split
Y = churn_dummy[['Exited']]
X = churn_dummy.drop(columns=['Exited'])

```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)

len(x_train), len(x_test), len(y_train), len(y_test)
```

# Building the ANN

## step 1. Initializing the ANN

```python
ann = tf.keras.models.Sequential()   # ANN initialize 
```

## step 2. Adding the input layer and first hidden layer


```python
ann.add(tf.keras.layers.Dense(units= 6, activation='relu'))  # unit = no of input; acticvation: input paactivation function  using 'relu'
```

## step 3. Adding the second hidden layer


```python
ann.add(tf.keras.layers.Dense(units= 6, activation='relu'))
```

## step 4. Adding the  output layer


```python
ann.add(tf.keras.layers.Dense(units= 1, activation='sigmoid'))   # here unit layer = n-1; 
```

# Compiling the ANN

```python
ann.compile(optimizer = 'adam', loss='binary_crossentropy', metrics = 'accuracy')
```

# Training the ANN

```python
ann.fit(x_train, y_train, batch_size=32, epochs= 100)
```

```python
y_test
```

```python
y_test['Prediction'] = ann.predict(x_test)
```
